```python
import pandas as pd 
from glob import glob 
import json 
import numpy as np 
```

```python
jsons = glob("*.json")
```

```python
jsons_filtered = list(filter(lambda x: False if x.endswith("optim.py_data.json") else True,jsons))
index = [str(item.split("_")[0]).upper() + " " +  str(item.split("_")[1]).upper() for item in jsons_filtered]
print([item.endswith(".json") for item in jsons])
print(index)
df = pd.DataFrame(data=np.zeros((9,5)),columns=["Original Mean CCC", "Tested Valence CCC", "Tested Arousal CCC", "Tested Dominance CCC", "Trained Mean CCC"],index=index)
df
```

```python
for file in jsons_filtered:
    with open(file) as f:
        test_idx = int(jsons_filtered.index(file))
        data = json.load(f)
    df.at[index[test_idx],"Original Mean CCC"] = round(data["First Eval"],3)
    df.at[index[test_idx],"Tested Valence CCC"] = round(data["Second Eval whole"][0],3)
    df.at[index[test_idx],"Tested Arousal CCC"] = round(data["Second Eval whole"][1],3)
    df.at[index[test_idx],"Tested Dominance CCC"] = round(data["Second Eval whole"][1],3)
    df.at[index[test_idx],"Trained Mean CCC"] = round(data["Third Eval"],3)
df
```
