```python
from email.mime import audio
from glob import glob 
import numpy as np 
from random import sample
import wave
from glob import glob
from scipy.io import wavfile
from scipy.signal import resample

def extract_no_from_str(string):
    s = ""
    for i in string:
        if i.isdigit():
            s += i 
    return int(s) 

audio_files = glob("test_data\output_repeated_splits_test\WAVs\*.wav")
audio_files = [item.split('\\')[-1] for item in audio_files]
# FeatureSet.eGeMAPSv01b FeatureLevel.LowLevelDescriptors (496, 23)

container = np.empty(shape=(47,))
dialogues = []

for item in audio_files:
    dia_and_utt = (item.split("_")[-2:])
    dialogues.append([extract_no_from_str(word) for word in dia_and_utt])

dialogues = np.array(dialogues)
utts = dialogues[:,1]
dias = dialogues[:,0]
```

```python
import pandas as pd 
df1 = pd.read_csv('test_sent_emo.csv')
print(df1["Emotion"].value_counts())
print(df1["Speaker"].value_counts())
print(df1.shape)
```

```python
d = {}
for i in np.unique(dias):
    i_idx = dias == i
    a = np.unique(utts[i_idx]).tolist()
    d[str(i)] = a
    del(a)
d
```

```python
u_dias = np.unique(dias)
for i in range(df1.shape[0]):
    dia_id = df1.at[df1.index[i],"Dialogue_ID"]
    utr_id =  df1.at[df1.index[i],"Utterance_ID"]
    df1.at[df1.index[i],"Valid"] = dia_id in u_dias and utr_id in d[str(df1.at[df1.index[i],"Dialogue_ID"])]

audio_files
df1_valid = df1[df1["Valid"]]
```

```python
df1_valid
```

```python
category_to_dimension_valence = {"neutral":0,"joy":0.76,"anger":-0.43,"surprise":0.4,"sadness":-0.63,"disgust":-0.6,"fear":-0.64}
category_to_dimension_arousal = {"neutral":0,"joy":0.48,"anger":0.67,"surprise":0.67,"sadness":0.27,"disgust":0.35,"fear":0.6}
category_to_dimension_dominance = {"neutral":0,"joy":0.35,"anger":0.34,"surprise":-0.13,"sadness":-0.33,"disgust":0.11,"fear":-0.43}
```

```python
df1_valid["Valence"] = df1_valid["Emotion"].apply(lambda x: category_to_dimension_valence[x])
df1_valid["Arousal"] = df1_valid["Emotion"].apply(lambda x: category_to_dimension_arousal[x])
df1_valid["Dominance"] = df1_valid["Emotion"].apply(lambda x: category_to_dimension_dominance[x])
df1_valid
```

```python
print(len(audio_files))
audio_slice = df1_valid[["Dialogue_ID","Utterance_ID"]].to_numpy().tolist()
audio_slice_idx = [item in audio_slice for item in dialogues.tolist()]
print("LEN AUDIO SLICE:",len(audio_slice_idx))
audio_left = np.array(audio_files)[audio_slice_idx]
print(len(audio_left))
audio_left.tolist()
```

```python
new_audio_files_in_label_order = []
for i in range(df1.shape[0]):
    dia_id = df1.at[df1.index[i],"Dialogue_ID"]
    utr_id =  df1.at[df1.index[i],"Utterance_ID"]
    name = "dia" + str(dia_id) + "_utt" + str(utr_id) + ".wav"
    new_audio_files_in_label_order.append(name)

len(new_audio_files_in_label_order)
```

```python
df1_valid_without_neutral = df1_valid[~(df1_valid["Emotion"] == "neutral")]

new_audio_files_in_label_order_no_neutral = []
for i in range(df1_valid_without_neutral.shape[0]):
    dia_id = df1_valid_without_neutral.at[df1_valid_without_neutral.index[i],"Dialogue_ID"]
    utr_id = df1_valid_without_neutral.at[df1_valid_without_neutral.index[i],"Utterance_ID"]
    name = "dia" + str(dia_id) + "_utt" + str(utr_id) + ".wav"
    new_audio_files_in_label_order_no_neutral.append(name)

len(new_audio_files_in_label_order_no_neutral)
```
